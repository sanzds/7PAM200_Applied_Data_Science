#Line plot

```python
# 1. Select Data
subdata = df[['date','emp_prop','hsemp_prop','unemp_prop', 'graduate_type', 'age_group']]  # Selecting data for first visualisation
subdata = subdata[subdata['age_group'] == 'Age Group 16-64'] #Selecting data for people of all ages
#Making catogory's line
a = subdata[subdata['graduate_type'] == 'Postgraduate']
b = subdata[subdata['graduate_type'] == 'Non-Graduate']
c = subdata[subdata['graduate_type'] == 'Graduate']
# 2. Setup Plot
fig, ax = plt.subplots(figsize = (25, 7)) #width, height
# 3. Plot data
plt.plot(a.date,a.emp_prop, label = "Post-Graduate", )
plt.plot(b.date,b.emp_prop, label = "Non-Graduate")
plt.plot(c.date,c.emp_prop, label = "Graduate")

ax.set(title="Employment rate by type of Graduation in Uk",
xlabel="Date",
ylabel="Employment Proability(%)")

plt.legend()
plt.show()

```

# Grouped Bar Chart

```python
x=None
subdata = None
a=None
b=None
c =None
subdata = df2[["time_period","age_group","graduate_type","gender","median"]]  # Selecting data for first visualisation
subdata = subdata[subdata['age_group'] == 'Age Group 16-64'] #Selecting data for people of all 
subdata = subdata[subdata['gender'] == 'Total']
a = subdata[subdata['graduate_type'] == 'Postgraduate']
b = subdata[subdata['graduate_type'] == 'Graduate']
c = subdata[subdata['graduate_type'] == 'Non-Graduate']
#Making lists of each type and label
labels = a.time_period
post = a["median"]
non = c["median"]
grad = b["median"]
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(19,7))
rects1 = ax.bar(x - width/2, post, width, label='PostGraduate')
rects3 = ax.bar(x, grad, width, label='Graduate')
rects2 = ax.bar(x + width/2, non, width, label='Non-Graduate')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Median Salary(Pounds)')
ax.set_title('Median Salary for Various Graduate types')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=2)
ax.bar_label(rects2, padding=2)
ax.bar_label(rects3, padding=2)

fig.tight_layout()

plt.show()
```

#Pie Chart

```python
subdata=df2.tail(6)
```

```python
subdata
```

```python
labels= ["Male-Graduate","Male-NonGraduate","Male-Postgraduate","Female-Graduate","Female-NonGraduate","Female-Postgraduate"]
data = subdata["median"] 
# Creating explode data
explode = (0.1, 0.0, 0.2, 0.3, 0.0, 0.0)
 
# Creating color parameters
colors = ( "orange", "cyan", "yellow",
          "magenta", "lightblue", "lightgreen")
 
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
 
# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} Pounds)".format(pct, absolute)
 
# Creating plot
fig, ax = plt.subplots(figsize =(10, 10))
wedges, texts, autotexts = ax.pie(data,
                                  autopct = lambda pct: func(pct, data),
                                  explode = explode,
                                  labels = labels,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="black"))
 
# Adding legend
ax.legend(wedges, labels,
          title ="Cars",
          loc ="upper center",
          bbox_to_anchor =(1, 0, 0.5, 1))
 
plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Median Salary of Various Graduate of Male and Female in 2021 (Pounds)")
 
# show plot
plt.show()
```
